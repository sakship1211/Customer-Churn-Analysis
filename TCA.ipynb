```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of customer by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

#from the given pie chart we can conclude that 26.54% of our customers have churned out. #not let's explore the reason behind it

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen ")
plt.show()
```

```python


# Calculate percentages
grouped = df.groupby(["SeniorCitizen", "Churn"]).size().unstack(fill_value=0)
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Define custom colors
colors = ["#1f77b4", "#ff7f0e"]

# Plot the stacked bar chart
percentages.plot(kind='bar', stacked=True, figsize=(5, 5), color=colors)

# Add percentage labels
for i, senior in enumerate(percentages.index):
    bottom = 0
    for churn in percentages.columns:
        value = percentages.loc[senior, churn]
        plt.text(
            i, bottom + value / 2, f'{value:.1f}%', ha='center', va='center', color='white'
        )
        bottom += value

# Customize the plot
plt.title("Churn by Senior Citizen (Stacked Bar Chart)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", bbox_to_anchor = (0.9,0.9))
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python


# List of columns
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Determine the layout of subplots
num_columns = len(columns)
rows = (num_columns + 2) // 3  # 3 plots per row

# Create subplots
fig, axes = plt.subplots(rows, 3, figsize=(15, rows * 4))  # Adjust figsize as needed
axes = axes.flatten()  # Flatten the axes for easy iteration

# Plot each column as a countplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f"Countplot of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

# Remove any unused subplots if columns < rows*3
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout for better appearance
plt.tight_layout()
plt.show()
```


 
#Customers using "Fiber optic" internet and those lacking add-on services like OnlineSecurity or TechSupport havenoticeably higher churn rates. 
In contrast, customers with PhoneService or Streaming services (TV and Movies) exhibit more balanced churn rates between "Yes" and "No." Notably, customers with "No internet service" consistently have very low churn rates across all categories.
